load the data analysis libraries

```python
import numpy as np 
import matplotlib.pyplot as plt
```

get the data/generate data

```python
rainfall = [17,10,10,10,26,59]
months = ['jan', 'feb', 'mar', 'apr', 'may', 'jun' ]
```

```python
plt.style.available
```

```python
plt.style.use('ggplot') # theme
```

visualize the data

```python
plt.bar(months, rainfall, color='#f90129')
```

```python
plt.pie(
    rainfall,
    labels=months,
    autopct = '%1.1f%%',
    explode=(0,0,0,0,0,0.1)
)
plt.show()
```

```python
plt.pie(
    rainfall,
    labels=months,
    autopct = '%1.1f%%',
)
plt.show()
```

```python

plt.pie(
    rainfall,
    labels=months,
    autopct = '%1.1f%%',
    wedgeprops={'width':0.4},
    pctdistance=.8
)
plt.show()
```

```python

plt.pie(
    rainfall,
    labels=months,
    autopct = '%1.1f%%',
    wedgeprops={'width':0.4,'edgecolor':'#6666ff','linewidth':2},
    pctdistance=.8
)
plt.show()
```

box plot

```python
plt.boxplot(rainfall,
            vert=False,
            showmeans=True,
            meanline=True,
            meanprops={'color': 'red'}
            )
plt.show()
```
